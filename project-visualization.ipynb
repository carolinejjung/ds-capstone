```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
dir_path = '/Users/carolinejung/Library/Mobile Documents/com~apple~CloudDocs/Desktop/8th sem SPRING 2025/DS 340H/Datasets/'
```

```python
df = pd.read_csv(dir_path+"bluebikedata/bluebikes_finalsubsetv1.csv")
df["trip_length_logged"] = np.log(df.trip_length)
```

```python
sns.barplot(data=df, x="round_trip", y="trip_length_logged", 
            hue="member_casual")
plt.suptitle("Mean Logged Trip Length based on Round Trip Rides and Membership Status")
plt.show()
```
